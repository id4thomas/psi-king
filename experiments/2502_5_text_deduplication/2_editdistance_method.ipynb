```python
import editdistance
import time

def deduplicate_strings(strings, length_threshold = 100, threshold=1):
    if len(strings)<2:
        return strings
    unique_strings = [strings[0]]
    for s in strings[1:]:
        is_duplicate = False
        for u in unique_strings:
            # print(editdistance.eval(s, u), threshold,  editdistance.eval(s, u) <= threshold, s, u)
            if editdistance.eval(s[:length_threshold], u[:length_threshold]) <= threshold:
                is_duplicate = True
                break
        if not is_duplicate:
            unique_strings.append(s)
    return unique_strings
```

```python
strings = [
    "hello"*1000,
    "hallo"*1000,
    "hello"*1000,
    "world"*1000,
    "wurld"*1000,
    "worlds"*1000
]
# editdistance <=threshold will be considered 
start = time.time()
deduped = deduplicate_strings(strings, threshold=3)
end = time.time()
print(f"{end-start:.3f}",len(deduped))
```
